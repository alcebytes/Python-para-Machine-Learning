## **Projeto** - Gráfico de regiões de probabilidade da distribuição normal

Vamos pausar por um momento e voltar ao scipy. Qual que é a probabilidade da região entre múltiplos de um desvio-padrão da distribuição normal padrão? Isto é
$$\mbox{Se }X \sim N(\mu, \sigma), \mbox{ então qual é o valor de }P(|X - \mu| \leq k\sigma),\mbox{ para } k = 1, 2, 3?$$

Sabemos que esse problema é equivalente ao seguinte
$$\mbox{Se }Z \sim N(0, 1), \mbox{ então qual é o valor de }P(-k \leq Z \leq k),\mbox{ para } k = 1, 2, 3?$$

```
p1 = ss.norm.cdf(1) - ss.norm.cdf(-1)
p2 = ss.norm.cdf(2) - ss.norm.cdf(-2)
p3 = ss.norm.cdf(3) - ss.norm.cdf(-3)
print('P(|X| < 1) = {}'.format(round(p1, 4)))
print('P(|X| < 2) = {}'.format(round(p2, 4)))
print('P(|X| < 3) = {}'.format(round(p3, 4)))
```
